```python
import math
a=int(input("Entrez la valeur du premier nombre"))
b=int(input("Entrez la valeur du second nombre"))
c=int(input("Entrez la valeur du troisième nombre"))
if a!=0:
    delta= b**2 -4*a*c
    if delta<0:
        print(f"l'équation {a}x² + {abs(b)}x +{c:.0f}  n'a pas de solution")
 
    elif delta==0:
        x=(-b)/(2*a)
        print(f"l'équation {a:.0f}x² + {abs(b)}x +{c:.0f} a pour  solution: {x:.0f} ")
    elif delta>0:
        x1=(-b-(delta**(0.5)))/2*a
        x2=(-b+(delta**(0.5)))/2*a
        print(f"les solutions possibles de l'équation {a:.0f}x² + {abs(b)}x +{c:.0f} sont: {x1:.2f} et {x2:.2f}")
else:
    print("Erreur la variable a doit être non nul")
```
